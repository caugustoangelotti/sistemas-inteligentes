```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
pd.set_option('display.max_rows', 500)
```

```python
df = pd.read_csv("happiness_treino.csv")
```

```python
df
```

```python
df.rename(columns={
    "Country or region": "region", 
    "GDP per capita" : "GDP",
    "Social support" : "social_support",
    "Healthy life expectancy" : "life_expectancy",
    "Freedom to make life choices" : "freedom",
    "Generosity" : "generosity",
    "Perceptions of corruption" : "perceptions_of_corruption"
}, inplace = True)
```

```python
df
```

```python
df.describe()
```

```python
X = np.array(df.drop("region", axis = 1))
```

```python
import sklearn.cluster as cluster
```

```python
kmeans = cluster.KMeans(n_clusters=10, init = "k-means++")
```

```python
kmeans = kmeans.fit(X)
```

```python
sns.scatterplot(X[:, 0], X[:, 1])
sns.scatterplot(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 80)
```

```python
df['Grupo'] = kmeans.labels_
```

# Classificação por pais 

```python
df
```

```python
df_teste = pd.read_csv("happiness_teste.csv")
```

```python
df_teste.rename(columns={
    "Country or region": "region", 
    "GDP per capita" : "GDP",
    "Social support" : "social_support",
    "Healthy life expectancy" : "life_expectancy",
    "Freedom to make life choices" : "freedom",
    "Generosity" : "generosity",
    "Perceptions of corruption" : "perceptions_of_corruption"
}, inplace = True)
```

```python
Y = np.array(df_teste.drop("region", axis = 1))
```

```python
kmeans = kmeans.predict(Y) 
```

# Grupos Semelhantes Aos Casos de Teste

```python
df_teste['Grupo'] = kmeans
df_teste
```

```python
dadosFinais = pd.concat([df, df_teste])
```
